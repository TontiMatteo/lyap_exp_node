```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 100)
y = np.maximum(0, x)
```

```python
plt.plot(x, y, color='blue', linewidth=2)
plt.title("ReLU function")
plt.xlabel("x")
plt.ylabel("S(x)")
plt.ylim(-3, 3)
plt.xlim(-3, 3)
# plt.axhline(y=0.5, color='gray', linestyle='--') # Adds a line at the midpoint
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-3, 3, 200)
alpha = 0.1

y = np.where(x >= 0, x, alpha * x)

plt.plot(x, y)
plt.title("Leaky ReLU")
plt.xlim(-3, 3)
plt.ylim(-3, 3)

plt.xlabel("x")
plt.ylabel("S(x)")
plt.grid(True)

plt.show()
```
